## Confidence Interval

Confidence interval is the term that depends on sample and standard error or Sampling error. It is hard to provide an estimate for a huge population, so instead we select a sample and try to provide an estimate of the measured parameter for that sample. This is called sample statistical inferencing. However, by taking different samples, the estimate will vary for each sample. Confidence interval is the interval to describe the estimate of that parameter in a range ultiamtely for a whole population. 

### Boostrap Principle

Boostrap principle is nothing but the process to use sample standard deviation and sample parameter instead that of the population, because we would not have those values for the population as we trying to make an estimate with confidence interval. 

Standard error is given by $$SE = \frac {σ}{\sqrt[]{n}} $$ 

and $$ {σ} = \sqrt{p(1-p)}$$

where p is the sample parameter value. 

Then we simply add and subtract the value to actual parameter estimate. 

$$ Actual Parameter Estimate+- z \frac {σ}{\sqrt[]{n}}$$

## Hypothesis Testing

We have to assume a null hypothesis which will be the contrary of the actual hypothesis we are making. Then we make tests on both and compare the hypothesis. If the actual hypothesis is true, we reject the null hypothesis.
