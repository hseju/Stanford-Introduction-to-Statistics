### Corelation Coefficient

Corelation coefficient describes the relantionship between the two variables namely, exploratory vairable on x axis and response variable on y axis. The corelation coefficient r ranges between -1 to 1. Consider the examples in the image for the different values of r. The higher the value of r suggests high linear corelation between the two variables and the scatter points are closer. Lower value of r suggests low linear corelation and scatter points are sparsed.  

It's important to note that the corelation coefficient is only given for linearl regression lines. It is not given for curved regression lines or polynomial curves. The value of r is given by equation:

$$ r = \frac{S_{x_y}} {{S_x}{S_y}} $$


where Sxy is the sample covariance and Sx and Sy are sample standard deviations. 


**NOTE: Corelations does not mean Causation**

## Regression Effect

In a test-retest situation virtually the bottom group on the first test will show some improvemet in the retest and the top group on first test will show degradation in retest. This is called Regression effect.

Predicting y from x in a equation of line given by y = mx +b is not the same as predicting x from y in the same equation. But rather to do this the two vaiables need to be flipped on the plot having the target variable on y axis. 

### Residual Plot:

The residula plot is the plot generated by calculating the difference in the original y value and predicted y value. The residual plot gives an idea about the linear regression line usefulness and accuracy. From the residual plot we can also get to know if we need to transform the data to correctly developed a regression line. 

Residual plots can not be useful when we have outliers in the data. The outliers can heavily affect the regression line and hence the residula plot as well.
